## 构建房价预测模型

### 导入模块

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
df
```

### 特征工程

```python
df['市区'].unique()
```

```python
# 删除所有缺失值
d1 = df.dropna().reset_index(drop=True)

# 编码-有序多分类（根据上面可视化的结果，按照对价格的影响程度排序，越大影响越高）
# 无序多分类无法直接引入，必须“哑元”化变量
# 等级变量（有序多分类）可以直接引入模型
map1 = {'南':5, '南北':6, '北':1, '西南':10, '东西':4, '东':2, '东北':8, '东南':9, '西':3, '西北':7}
d1['朝向'] = d1['朝向'].map(map1)
map2 = {'毛坯':1, '简装':2, '其他':3, '精装':4}
d1['装修情况'] = d1['装修情况'].map(map2)
map3 = {'有电梯':1, '无电梯':0,'未知':2}
d1['电梯'] = d1['电梯'].map(map3)
map4 = {'朝阳':1,'海淀':2,'丰台':3,'昌平':4,'大兴':5,'房山':6,'平谷':7,'怀柔':8,'密云':9,'门头沟':10,'顺义':11,'延庆':12,'通州':13,'石景山':14,'西城':15,'东城':16}
d1['市区'] = d1['市区'].map(map4)

# 删除超过2010年的房子，年代转变为房龄
d1['年份'] = d1['年份'].astype('int32')
d1.drop(index=d1[d1['年份']>2010].index,inplace=True)
d1['房龄'] = d1['年份'].map(lambda x: 2022-x)

# 删除超过200平的房子
d1['面积(㎡)'] = d1['面积(㎡)'].astype('int32')
d1.drop(index=d1[d1['面积(㎡)']>200].index,inplace=True)

d1.drop(columns=['年份','小区','户型','价格(万元)'],inplace=True)
```

```python
d1
```

### knn回归算法模型

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error,mean_absolute_error

# 导入二手房数据
data = d1[['市区','朝向','楼层','装修情况','电梯','面积(㎡)','房龄','单价(万元)']]
X = data.iloc[:,:-1]
y = data.iloc[:,-1]

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)

# 创建一个线性回归模型
Knn = KNeighborsRegressor()

# # 利用gscv调整n_neighbors参数
# # 创建参数字典
# param_dict = {'n_neighbors':[x for x in range(1,51)]}
# # step 1:实例化gscv对象
# gscv = GridSearchCV(estimator=Knn,param_grid=param_dict,cv=5)
# # step 2:训练
# gscv.fit(x_train,y_train)
# # step 3:获取gscv结果
# gscv.best_params_
# # 输出：{'n_neighbors': 7}
```

```python
Knn.fit(x_train,y_train)
```

```python
y_pre_Knn = Knn.predict(x_test)
```

```python
round(mean_absolute_error(y_test,y_pre_Knn),2)
```

```python
round(mean_squared_error(y_test,y_pre_Knn),2)
```

```python
Knn.score(x_test,y_test)
```

### 将所有模型及其评估标准对比输出

```python
models=[KNeighborsRegressor(n_neighbors=7),LinearRegression(),Ridge(),Lasso(),DecisionTreeRegressor(),RandomForestRegressor()]
models_name=['KNNRegressor','LinearRegression','Ridge','Lasso','DecisionTree','RandomForest']
score_=[]
```

```python
for name,model in zip(models_name,models):
    print('开始训练模型：'+name)
    model=model   #建立模型
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)  
    score=model.score(x_test,y_test)
    score_.append(str(score)[:5])
    print(name +' 得分:'+str(score))
```
